```python
# T-test (independent sample t-test)


# By running the test, we're looking at data from both departments and
# deciding if the difference is significant enough to conclude that
# the job satisfaction is genuinely different between the two groups,
# or if it’s small enough that it could have just happened randomly.


# TO TEST:
# the (job-satis)ratings of sales dept are diff from prod-dev dept with actual reason , or diff just by chance(Randomnly)
# if the diff has actual reason, then what might be that reason is the INSIGHT..

# WHAT WE GET ?
# If the average job satisfaction score in the Sales department is 4.2 and in the prod-dev department is 3.6,
# we might want to know: Is that difference real, or could it just be due to random chance?

# There’s something about working in Sales that makes people less satisfied than in prod-dev.
# This difference is real, not just something that happened randomly when we collected data.



# Null Hypothesis (H0):
#  There is no difference in job satisfaction between Product Development and Sales departments.

# Alternative Hypothesis (H1):
# There is a difference in job satisfaction between Product Development and Sales departments.

```

```python
from scipy import stats

# Extract Job Satisfaction data for Product Development and Sales departments
prod_dev_satisfaction = df[df['Department'] == 'Product Development']['Job Satisfaction']
sales_satisfaction = df[df['Department'] == 'Sales']['Job Satisfaction']

# Perform the independent t-test
t_stat, p_value = stats.ttest_ind(prod_dev_satisfaction, sales_satisfaction, equal_var=False)

# Print the t-statistic and p-value
print(f"T-statistic: {t_stat}, P-value: {p_value}")

# Decision rule based on p-value
if p_value < 0.05:
    print("Reject the null hypothesis: There is a significant difference in job satisfaction between Product Development and Sales.")
else:
    print("Fail to reject the null hypothesis: No significant difference in job satisfaction between Product Development and Sales.")
```

```python
# RESULTS AND CONCLUSION:


# Since the t-statistic is very close to 0 (in this case, -0.083),
# it indicates that the difference between the two groups is almost negligible.

# the p-value is 0.937, which is much higher than 0.05.
# This means the difference in job satisfaction between the Product Development and Sales departments
# is likely due to random variation rather than any actual difference in job satisfaction.

# Fail to reject the null hypothesis: This means that based on the data,
# we do not have enough evidence to say there is a statistically significant difference in job satisfaction between employees in the Product Development and Sales departments.

# In simple terms:
# The job satisfaction levels between the two departments are very similar,
# and the small differences we observe are likely just due to RANDOM CHANCE.


# THAT IS,

# Imagine you flip two coins,
# one from Product Development and one from Sales, 100 times.
# Both coins seem to land on heads about 50 times,
# so the results are so close that you conclude both coins are likely fair.
# Similarly, your test tells you that the job satisfaction in both departments is about the same.
# There's no evidence to suggest that one department is more or less satisfied than the other.

```

```python
# ---- UNDERSTAND T-TEST ----

# If you had two classes in school:

# Class A has an average test score of 85.
# Class B has an average test score of 90.
# If the difference is statistically significant,
# it means that the 5-point difference isn’t just random—Class B really does tend to score higher than Class A.

# if accept null hypothese, then there might a reson of diff like,
# the reason y class B has 5 marks more might be,
# - extra classes
# - good faculty
# - more hardwork by students

#  if reject null hypothesis, then there is no reson in diff betwn them
```

```python
# ANOVA TEST - 1 way


# on departments : Sales, HR, and Product Development.
# to determine : if job satisfaction is significantly different across these departments.
# helps answer the question:  Are employees in some departments more satisfied than others?
# "Do employees in Sales have higher job satisfaction than those in HR or Product Development?"
# "Is one department significantly different in how satisfied their employees are compared to others?"


# F-value: ratio of variation between group means to the variation within the groups

# The P-value helps determine whether the observed differences between group means are statistically significant.

# - A small P-value (typically < 0.05)
#   means that the differences between the groups are statistically significant,
#   and you can reject the null hypothesis (which says that the group means are equal).
# - A large P-value (typically ≥ 0.05)
#   suggests that the differences between the groups are likely due to random chance,
#   so you fail to reject the null hypothesis.

# Post-hoc tests (if needed):
#   If ANOVA finds a significant difference, you can then perform post-hoc tests (e.g., Tukey’s test)
#   to find out which specific pairs of departments (A vs. B, B vs. C, etc.) are different.


import scipy.stats as stats

# Grouping Job Satisfaction based on Department
sales_satisfaction = df[df['Department'] == 'Sales']['Job Satisfaction']
hr_satisfaction = df[df['Department'] == 'HR']['Job Satisfaction']
proddev_satisfaction = df[df['Department'] == 'Product Development']['Job Satisfaction']

# Performing ANOVA
fvalue, pvalue = stats.f_oneway(sales_satisfaction, hr_satisfaction, proddev_satisfaction)

print(f"F-value: {fvalue}, P-value: {pvalue}")

# Decision rule based on p-value
if pvalue < 0.05:
    print("Reject the null hypothesis: There is a significant difference in job satisfaction between departments.")
else:
    print("Fail to reject the null hypothesis: No significant difference in job satisfaction between departments.")

```

```python
# RESULTS AND CONCLUSION :


# F-value: 0.678 suggests that the variation between the job satisfaction means of Sales, HR, and Product Development isn't large.
# P-value: 0.542 (fail to reject NULL HYPOTHESIS)
# means there is no statistically significant difference in job satisfaction between the departments.
# The observed differences are likely due to random chance rather than any real difference in job satisfaction between departments.


# In other words,
# the job satisfaction levels across these departments are likely to be similar,
# with no strong evidence showing one department is more or less satisfied than the others.
```

```python
# Chi-Square test

# to determine if there is a significant association between categorical variables in your dataset
# we have 2 categorical varables and we want to test whether there is a relationship, we use chi-square test
# ex) is there a relation between gender(male and female) and preferred news paper(hindu, TOI, eenadu)

# H0: there is no relationship btwn them
# H1: there is a relation btwn them
```

```python
# Grouping by Age Bracket to Job Satisfaction
age_satisfaction = df.groupby('Age Bracket')['Job Satisfaction'].mean()
print("Job Satisfaction by Age Bracket:")
print(age_satisfaction)

# These scores give you a general sense of which age group tends to have higher or lower job satisfaction
# based on numerical values. The higher the score, the more satisfied the group is on average.
# From this, you can infer that the 18-24 age group seems more satisfied on average compared to others.

# However, this is a descriptive analysis—it does not tell you whether the differences between groups
#  are statistically significant or due to random variation.
```

```python
# The chi-square test doesn't deal with averages. Instead, it focuses on categorical data and tests
# whether there’s a significant association between two categorical variables


# DIFFERNECE BTWN average job satisfaction scores (like ABOVE) and the chi-square test :

# Averages give you a summary (higher score means more satisfaction), but do not test if differences are significant.
# The chi-square test evaluates whether the differences in satisfaction (based on response categories)
# between age groups are statistically significant.
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Create a contingency table of counts
contingency_table = pd.crosstab(df['Age Bracket'], df['Job Satisfaction'])

print("Contingency Table:")
print(contingency_table)

# Perform the Chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Output the results
print(f"\nChi-square Statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of Freedom: {dof}")
print("\nExpected Frequencies:")
print(expected)

# Interpretation
if p < 0.05:
    print("\nThere is a significant association between Age Bracket and Job Satisfaction.") # reject null hypotheses
else:
    print("\nThere is no significant association between Age Bracket and Job Satisfaction.") # fail to reject null hypothesis
```

```python
# The observed differences in job satisfaction across different age groups are likely due to chance.

# 2nd method to test hypothesis is by comparing with CRITICAL CHI-SQUARE VALUE

# If your chi-square statistic > critical value, you REJECT the null hypothesis.
# If it's less, you FAIL TO REJECT the null hypothesis.


# https://datatab.net/tutorial/chi-square-distribution  -- Table of chi-squared distribution
# ( to get critical chi square value)
# critical chi-square vlaue = 21.026 (by table)

# or -- by code

from scipy.stats import chi2

dof = 12  # DoF=(number of rows−1)×(number of columns−1)
alpha = 0.05
# Your chi-square statistic from your previous result
chi_square_statistic = 17.92

# Find the critical value
critical_value = chi2.ppf(1 - alpha, dof)

print(f"Critical Chi-Square Value: {critical_value}")

# Decision based on comparison of chi-square statistic and critical value
if chi_square_statistic > critical_value:
    print("\nReject the null hypothesis (H0)")
    print("There is a significant association between Age Bracket and Job Satisfaction.")
else:
    print("\nFail to reject the null hypothesis (H0)")
    print("There is no significant association between Age Bracket and Job Satisfaction.")
```
